```python
import pandas as pd
import numpy as np

# find the path to the data
path = 'pandas_intro/data'
```

```python
# load the data for exercise 5-6
college = pd.read_csv( f'{path}/college.csv')
college.head()
```

## Exercise 5. Find the mean and standard deviation of math and verbal SAT score for men-only, women-only, and non gender specific universities.

```python
# engineer a new feature.
# 0 is non-gender specific universities
# 1 is menonly university
# 2 is womenonly university 
ind =[ 'non gender specific', 'menonly', 'womenonly']
college['uni_type'] = college.menonly * 1 + college.womenonly * 2
# group by new feature
res_mean = college.groupby('uni_type').agg({'satmtmid': 'mean'}).dropna()
res_mean.index = ind
res_std  = college.groupby('uni_type').agg({'satmtmid': 'std'}).dropna()
res_std.index = ind
print( f'mean :\n{res_mean}\n')
print( f'std :\n{res_std}')
```

## Exercise 6. Find the top 3 universities with largest numbers of undergraduate students for each state

```python
res_top3 = college.loc[ :, ['instnm', 'ugds', 'stabbr']].sort_values('ugds').groupby('stabbr').head(3)
print( res_top3.sort_values('stabbr'))
```

```python
# load the data for exercise 5-6
emp = pd.read_csv( f'{path}/employee.csv')
emp.head()
```

## Exercise 7. Generate a DataFrame for the ratios of (number of employees of specific gender and race/total number of employees) for all race-gender combinations. 

```python
emp = emp.dropna()
N = emp.shape[0] 
res = emp.groupby(['race', 'gender']).size() / N
print(res.round(3))
```

### Exercise 8. Use `pd.melt()` to unpivot table `pv3` to the format of `pv4` in exampel above.

```python
pv2 = emp.pivot_table(index='gender', columns='race', 
                      values='salary', aggfunc='mean').round(-3)
pv3 = pv2.reset_index()
pv3 
```

```python
pv4 = pv3.melt(id_vars=['gender']).sort_values('gender')
pv4
```
