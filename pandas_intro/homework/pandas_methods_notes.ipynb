T implementation with existing data. by doc DataFrame.T Transpose index and columns

```python
import pandas as pd

data = pd.read_csv('../data/college.csv')
origin_df = pd.DataFrame(data)
#origin data
print(origin_df.head(5))
# DataFrame.T test
test_df = origin_df.copy()
test_df = test_df.T
print(test_df.head(27))
```

DataFrame.asfreq() by doc Convert time series to specified frequency.

```python
import pandas as pd

index = pd.date_range('1/1/2000', periods=4, freq='T')
series = pd.Series([0.0, None, 2.0, 3.0], index=index)
df = pd.DataFrame({'s': series})
#df
df.asfreq(freq='30S')
```

DataFrame.clip by doc Trim values at input threshold(s). this means that min input and max input will be limite by clip method to dataframe 

```python
#here orgin df have more than 6 and less -4
data = {'col_0': [9, -3, 0, -1, 5], 'col_1': [-2, -7, 6, 8, -5]}
df = pd.DataFrame(data)
df
```

```python
#after clip the all more than 6 are chagned to 6 and less than -4 changed to -4
df.clip(-4, 6)
```
