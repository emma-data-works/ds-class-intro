### Exercise 4: Is there statistical significant difference between average `md_earn_wne_p10` (Median Earnings 10 years after enrollment) for schools in Texas and California? 

Hint: we can see this problem as a [two-sample t-test](https://online.stat.psu.edu/stat555/node/36/)


```python
import pandas as pd
from scipy.stats import ttest_ind

#import data from csv file
college_df =pd.read_csv('../data/college.csv')

#filter out by state name
filter_1 = college_df['stabbr'] == "TX"
filter_2 = college_df['stabbr'] == "CA"

#shape A group and B group up data that select only 'md_earn_wne_p10' , value is not "PrivacySuppressed" and replace null values to 0 
A_df = pd.to_numeric(college_df[filter_1 & (college_df['md_earn_wne_p10'] != "PrivacySuppressed")]['md_earn_wne_p10'].fillna(0))
B_df = pd.to_numeric(college_df[filter_2 & (college_df['md_earn_wne_p10'] != "PrivacySuppressed")]['md_earn_wne_p10'].fillna(0))

#Independent Two Sample t-Test
ttest_ind(A_df,B_df)

#calculate the Welchâ€™s t-Test in Pandas
#ttest_ind(A_df,B_df,equal_var=False)

```

### Exercise 5. Find the mean and standard deviation of math and verbal SAT score for men-only, women-only, and non gender specific universities.

```python
import pandas as pd

college_df =pd.read_csv('../data/college.csv')


#df.groupby('grouping column').agg({'aggregating column': 'aggregating function', ... ... })

onlywomen_df = college_df[college_df['womenonly'] == 1.0]
onlywomen_df.groupby('instnm').agg({'satmtmid':['mean','std']}).fillna(0)
```

```python
onlymen_df = college_df[college_df['menonly'] == 1.0]
onlywomen_df.groupby('instnm').agg({'satmtmid':['mean','std']}).fillna(0)
```

```python
non_gender_df = college_df[(college_df['womenonly'] == 0.0) & (college_df['menonly'] == 0.0)]
non_gender_df.groupby('instnm').agg({'satmtmid':['mean','std']}).fillna(0)
```

### Exercise 7. Generate a DataFrame for the ratios of (number of employees of specific gender and race/total number of employees) for all race-gender combinations. 

```python
import pandas as pd

emp_df =pd.read_csv('../data/employee.csv')
emp_df.head(10)
```

```python
total_num = len(emp_df.index)
race_group = emp_df['race'].drop_duplicates()

whiteMale = len(emp_df[(emp_df['race'] == 'White') & (emp_df['gender'] == 'Male')].index)
whiteFeMale = len(emp_df[(emp_df['race'] == 'White') & (emp_df['gender'] == 'Female')].index)

print(whiteMale/total_num,whiteFeMale/total_num)

BlackMale = len(emp_df[(emp_df['race'] == 'Black') & (emp_df['gender'] == 'Male')].index)
BlackFeMale = len(emp_df[(emp_df['race'] == 'Black') & (emp_df['gender'] == 'Female')].index)

print(BlackMale/total_num,BlackFeMale/total_num)

AsianMale = len(emp_df[(emp_df['race'] == 'Asian') & (emp_df['gender'] == 'Male')].index)
AsianFeMale = len(emp_df[(emp_df['race'] == 'Asian') & (emp_df['gender'] == 'Female')].index)

print(AsianMale/total_num,AsianFeMale/total_num)

NaNMale = len(emp_df[(emp_df['race'] == 'NaN') & (emp_df['gender'] == 'Male')].index)
NaNFeMale = len(emp_df[(emp_df['race'] == 'NaN') & (emp_df['gender'] == 'Female')].index)

print(NaNMale/total_num,NaNFeMale/total_num)

HispanicMale = len(emp_df[(emp_df['race'] == 'Hispanic') & (emp_df['gender'] == 'Male')].index)
HispanicFeMale = len(emp_df[(emp_df['race'] == 'Hispanic') & (emp_df['gender'] == 'Female')].index)

print(HispanicMale/total_num,HispanicFeMale/total_num)

NAMale = len(emp_df[(emp_df['race'] == 'Native American') & (emp_df['gender'] == 'Male')].index)
NAFeMale = len(emp_df[(emp_df['race'] == 'Native American') & (emp_df['gender'] == 'Female')].index)

print(NAMale/total_num,NAFeMale/total_num)
```
