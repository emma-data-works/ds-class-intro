```python
ca = pd.to_numeric(college.loc[college['stabbr']=='CA', 'md_earn_wne_p10'], errors='coerce').dropna()
tx = pd.to_numeric(college.loc[college['stabbr']=='TX', 'md_earn_wne_p10'], errors='coerce').dropna()
ca_n = len(ca)
tx_n = len(tx)
ca_mean = ca.mean()
tx_mean = tx.mean()
ca_var = ca.std() ** 2
ca_var = ca.std() ** 2
tx_var = tx.std() **2
```

2-sided two-sample mean test
- h0 : ca == tx, or delta == 0
- ha/h1 : ca != tx, or delta !=0

```python
# assuming unequal variance
delta = ca_mean - tx_mean
se = np.sqrt(ca_var / ca_n + tx_var / tx_n)

t_val = delta / se
t_val
```

```python
# 2-sided p-value
# using t-distribution  ~~ this validates well with the scipy function
t_df = ca_n + tx_n -2
st.t.sf(abs(t_val), t_df) * 2 # calculate p-value (both sides) using t distribution
```

```python
# using standard normal
st.norm.sf(abs(t_val)) * 2 # calculate p-value (both sides) using normal distribution
```

```python
# call scipy module function for two sample t-test
st.ttest_ind(ca, tx, equal_var=False)
```

1-sided two-sample mean test
- h0 : ca > tx, or delta > 0
- ha/h1 : ca <= tx, or delta <= 0

```python
# calculation is the same
# 1-sided p-value
# using t-distribution
st.t.sf(t_val, t_df) # calculate p-value (one side) using t distribution
```

```python
# using standard normal
st.norm.sf(t_val)# calculate p-value (one side) using normal distribution
```
