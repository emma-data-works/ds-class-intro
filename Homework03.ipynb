### Exercise 4: Is there statistical significant difference between average `md_earn_wne_p10` (Median Earnings 10 years after enrollment) for schools in Texas and California? 

Hint: we can see this problem as a [two-sample t-test](https://online.stat.psu.edu/stat555/node/36/)

#### you might find the following function useful to converts data to numeric
pd.to_numeric()

#### example code for p-value calculation if you'd like to calculate z-scores
import scipy.stats as st
st.norm.cdf(-1.506666)

#### you may also find the following function useful
st.ttest_ind()

```python
college[['md_earn_wne_p10']] = college[['md_earn_wne_p10']].apply(pd.to_numeric,errors="coerce")

CA = college.loc[college['stabbr']=='CA',['md_earn_wne_p10']].dropna().to_numpy()
TX = college.loc[college['stabbr']=='TX',['md_earn_wne_p10']].dropna().to_numpy()
```

```python
import scipy.stats as st
st.ttest_ind(a= CA,b= TX, equal_var= True, alternative='two-sided')
```

```python
import math
z_score = (CA.mean()-TX.mean())/math.sqrt(CA.std()**2/len(CA) +TX.std()**2/len(TX))
z_score
```

```python
st.norm.cdf(z_score)
```

```python
df_sat_women = college.loc[college['womenonly']==1,].groupby('womenonly').agg({'satvrmid':['mean','std'], 'satmtmid':['mean','std']})
df_sat_women
```

### Exercise 7. Generate a DataFrame for the ratios of (number of employees of specific gender and race/total number of employees) for all race-gender combinations. 

```python
emp = pd.read_csv('data/employee.csv')
emp.head()
```

```python
emp.pivot_table(index='gender',columns='race',aggfunc='size').div(len(emp)).mul(100).round(1)
```

### Exercise 8. Use `pd.melt()` to unpivot table `pv3` to the format of `pv4` in exampel above.

```python
pv3 = emp.pivot_table(index='gender', columns='race', 
                      values='salary', aggfunc='mean').round(-3).reset_index()
pv3
```

```python
pv4= pd.melt(pv3,id_vars='gender',value_name='salary')
pv4
```
